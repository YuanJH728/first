## 1、数据摘要和可视化

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#标称属性，给出每个可能取值的频数,由上面给出的数据类型，标称属性有repositories和language,但language相对有价值
print('------------------------------language可能取值频数------------------------------------')
df['language'].value_counts().head(10).plot.bar()
plt.title('language')
```

```python
#数值属性，给出5数概括及缺失值的个数
print('-------------------------------数据类型的五数概括和均值以及标准差-------------------------------')
print(df.describe())
print('--------------------------------缺失值的个数统计----------------------------------------------------')
print(df.isnull().sum())
```

### 分析
首先分析标称属性，从仓库名称看，一个仓库最多出现两次。再观察使用的编程语言，使用最多的是JavaScript,共有253个项目使用了这种编程语言。这些项目总共使用了51种编程语言。再分析数值型属性，从频率图看，上述五个属性的取值大多在0-100之间，第一四分位数，均值，中位数，第三四分位数都小于100，数据盒图可以验证这一点

## 2、数据缺失的处理

```python
#将缺失部分剔除
df2 = df.dropna(how = 'any')
#求出新数据五数概括
print(df2.describe())
#画出新数据盒图
label=['stars_count','forks_count','issues_count','pull_requests','contributors']
df3 = (df2['stars_count'],df2['forks_count'],df2['issues_count'],df2['pull_requests'],df2['contributors'])
plt.boxplot(df3,labels=label,meanline=True)
plt.show()
```

由于删除部分数据，新数据有所变化，比如stars_count和issue_count的最大值变小，数据总行数变少。但盒图却没有显著变化，可能是被删除数据分布均匀或者是删除数据较少不影响数据总体分布。

```python
#用最高频率值来填补缺失值,只有language属性有空值，且该属性中频率最高是JavaScript，故用该值取代空值
df2 = df.fillna(value='JavaScript')
#求出新数据五数概括
print(df2.describe())
#画出新数据盒图
label=['stars_count','forks_count','issues_count','pull_requests','contributors']
df3 = (df2['stars_count'],df2['forks_count'],df2['issues_count'],df2['pull_requests'],df2['contributors'])
plt.boxplot(df3,labels=label,meanline=True)
plt.show()
```

显然，只有标称属性有变化，数值属性没变，所以这不影响数据盒图和五数概括，会影响language属性的频率，即JavaScript的频率增加，而其他编程语言的频率不变。

```python
#通过属性的相关关系来填补缺失值,因为language是非数值型，无法计算属性相关性，所以这里通过通过数据对象之间的相似性来填补缺失值
df = pd.read_csv("github_dataset.csv")
k = 0
list = df['language'].isnull()
for x in df['language'].isnull():
    if(x == True):
        for y in range(1052):
            if list[y] == False:
                if np.absolute(df['stars_count'][k]-df['stars_count'][y])+np.absolute(df['forks_count'][k]-df['forks_count'][y])+np.absolute(df['issues_count'][k]-df['issues_count'][y])+np.absolute(df['pull_requests'][k]-df['pull_requests'][y])+np.absolute(df['contributors'][k]-df['contributors'][y])<750 :
                  df['language'][k] = df['language'][y] 
                  break
    k = k+1
print("language属性列空值个数：",df['language'].isnull().sum())#处理后空值个数
#求出新数据五数概括
print(df.describe())
#画出新数据盒图
label=['stars_count','forks_count','issues_count','pull_requests','contributors']
df3 = (df['stars_count'],df['forks_count'],df['issues_count'],df['pull_requests'],df['contributors'])
plt.boxplot(df3,labels=label,meanline=True)
plt.show()
#编程语言使用频率统计
df['language'].value_counts().head(10).plot.bar()
plt.title('language')
```

同样的，由于只改变标称属性，则数值属性的五数概括和盒图不变，language列的频率数将改变，使用数量最高的依然是JavaScript，且数量增加较少，而C++语言使用量增加很多，从29增加到139，位次排第三名，其他基本保持不变。

```python
# 通过属性的相关关系来填补缺失值
data = pd.read_csv('movies_dataset.csv')
data["downloads"] = data["downloads"].apply(pd.to_numeric, errors='coerce')
data["id"] = data["id"].apply(pd.to_numeric, errors='coerce')
data["views"] = data["views"].apply(pd.to_numeric, errors='coerce')
corr = data.corr()
#相关系数显示
print(corr)
#相关值替代
data["downloads"][np.isnan(data["downloads"])] = data["views"][np.isnan(data["downloads"])]
data["views"][np.isnan(data["views"])] = data["downloads"][np.isnan(data["views"])]
#恢复数据类型，object-int
data["downloads"] = data["downloads"].apply(pd.to_numeric, errors='coerce')
data["views"] = data["views"].apply(pd.to_numeric, errors='coerce')
#数值属性盒图
visit = pd.DataFrame(data['IMDb-rating'])
visit.plot.box()
plt.show()
visit = pd.DataFrame(data['downloads'])
visit.plot.box()
plt.show()
visit = pd.DataFrame(data['views'])
visit.plot.box()
plt.show()
#标称属性的每个可能取值的频数，这里展示频度前十
index = 1
plt.figure(figsize=(10,10), dpi=80).subplots_adjust(hspace=1)
plt.figure(1)
col = 2
row = 3
for attr in nominals:
    plt.subplot(row, col, index)
    index += 1
    freq = 10
    data[attr].value_counts().head(freq).plot.bar()
    plt.title(f'{attr}: frequency of top {freq}')
```

如果通过属性的相关关系来填补缺失值，经过观察修改过后标称属性频率图和数值属性盒图，可以看到此操作改变了原来数据集的分析，可能是由于属性间的相关关系不大造成的，另外由于该数据集有效数值属性少，并且缺失值往往伴随着该条实例同时缺失其它的属性值，因此该方法对于该数据集在准确性上做的不好。

```python
# 通过数据对象之间的相似性来填补缺失值
data = pd.read_csv('movies_dataset.csv', low_memory=False)
data["downloads"] = data["downloads"].apply(pd.to_numeric, errors='coerce')
data["views"] = data["views"].apply(pd.to_numeric, errors='coerce')
list = data.isnull()
for x in range(20547):
    if(list['downloads'][x] == True):
        idx = data["id"][x]
        for i in range(x,20547):
            if list['downloads'][i]==False and np.absolute(data["id"][i] - idx)<200000:
                data["downloads"][x] = data["downloads"][i]
                break
    if(list['views'][x] == True):
        idx = data["id"][x]
        for i in range(0,20547):
            if list['views'][i]==False:
                data["views"][x] = data["views"][i]
                break
#数值属性盒图
visit = pd.DataFrame(data['IMDb-rating'])
visit.plot.box()
plt.show()
visit = pd.DataFrame(data['downloads'])
visit.plot.box()
plt.show()
visit = pd.DataFrame(data['views'])
visit.plot.box()
plt.show()
#标称属性的每个可能取值的频数，这里展示频度前十
index = 1
plt.figure(figsize=(10,10), dpi=80).subplots_adjust(hspace=1)
plt.figure(1)
col = 2
row = 3
for attr in nominals:
    plt.subplot(row, col, index)
    index += 1
    freq = 10
    data[attr].value_counts().head(freq).plot.bar()
    plt.title(f'{attr}: frequency of top {freq}')
data.isnull().sum()
```

如果通过数据对象之间的相似性来填补缺失值，经过观察修改后各属性的直方图和盒图，可以看到数据有些变化，有数据丰富也有数据特性缺失，比如views值变得过于集中，这不能很好地体现数据本身的特异性。
